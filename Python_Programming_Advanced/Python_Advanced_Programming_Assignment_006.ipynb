# Question 1:

#### You are given two strings s and t. String t is generated by randomly shuffling string s and then adding one more letter at a random position. Return the letter that was added to t.

```python
def find_the_difference(s, t):
    
    missing_char = ''
    
    for char in t:
        if char not in s or t.count(char) > s.count(char):
            missing_char = char
            
    return missing_char

s = input('Enter the original string : ')
t = input('Enter the shuffled string with extra char : ')

print(f'\nAdded Character is : {find_the_difference(s,t)}')
```

# Question 2:

#### Given a function that accepts unlimited arguments, check and count how  many data types are in those arguments. Finally return the total in a list.

List order is:

[int, str, bool, list, tuple, dictionary]

```python
def count_datatypes(*args):
    
    data_types = [int, str, bool, list, tuple, dict]
    list_dtype = [type(ele) for ele in args]
    dtype_count = [list_dtype.count(ele) for ele in data_types]
    
    return dtype_count

print(f'List of count of data types is :',end = ' ')
print(f'{count_datatypes("Hello", "Bye", True, True, False, {"1": "One", "2": "Two"}, [1, 3], {"Brayan": 18}, 25, 23)}')
```

# Question 3:

#### A Fibonacci string is a precedence of the Fibonacci series. It works with any two characters of the English alphabet (as opposed to the numbers 0 and 1 in the Fibonacci series) as the initial items and concatenates them together as it progresses in a similar fashion as the Fibonacci series.

```python
def fib_str(size, list_):
    
    if size == 1:
        return list_[0]
    
    elif size == 2:
        return list_[1]
    
    else:
        return fib_str(size-1, list_) + fib_str(size-2, list_)
    
size = int(input('Enter the length of the fibonacci series : '))
list_ = input(f'Enter the initial 2 values of the fibinacci string series : ').split(',')

fib_str_series = []

for k in range(1,size+1):
    fib_str_series.append(fib_str(k, list_))
    
print(f'\nFibonacci string series of size {size} is {fib_str_series}')
```

#### Given an integer between 0 and 26, make a variable (self.answer). That variable would be assigned to a string in this format:

#### "nines:your answer, threes:your answer, ones:your answer"

You need to find out how many ones, threes, and nines it would at least take for the number of each to add up to the given integer when multiplied by one, three or nine (depends).


```python
def ones_threes_nines(num):
    
    dict_ = {'nines':0, 'threes':0, 'ones':0}
    
    if num >= 9:
        dict_['nines'] = num//9
        if num%9 >= 3:
            dict_['threes'] = (num-(9*int(num/9)))//3
            if (num-(9*int(num/9)))%3 != 0:
                dict_['ones'] = (num-(9*int(num/9))-(3*int((num-(9*int(num/9)))/3)))//1
        else:
            dict_['ones'] = (num-(9*int(num/9)))//1
            
    
    elif num >= 3:
        dict_['threes'] = (num)//3
        if (num-(3*int(num/3)))%3 != 0:
                dict_['ones'] = (num-(3*int(num/3)))//1
                
    else:
        dict_['ones'] = (num)//1
        
    return f"nines:{dict_['nines']}, threes:{dict_['threes']}, ones:{dict_['ones']}"

num = int(input('Enter the number : '))

print(f'\n{ones_threes_nines(num)}')
```

# Question 5:

#### Write a non-recursive function that takes an integer n and returns the Fibonacci sequence's value at index n.

```python
#As a non-recursive function

def fib(num):
    
    fib_series = [0,1]
    
    for k in range(2,num+1):
        fib_series.append(fib_series[k-1]+fib_series[k-2])
        
    return fib_series[-1]

num = int(input('Enter the fibonacci index : '))

print(f'\n{fib(num)}')
```
